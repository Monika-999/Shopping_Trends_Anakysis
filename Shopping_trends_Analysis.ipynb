```python
# @title Default title text
import pandas as pd
import random

# Define column names
columns = [
    "Customer S.No", "Age", "Gender", "Item Purchased", "Category", "Purchase Amount", "Location",
    "Size", "Color", "Season", "Review Rating", "Subscription Status", "Payment Method",
    "Shipping Type", "Discount Applied", "Promo Code Used", "Previous Purchase",
    "Preferred Payment Method", "Frequency of Purchase"
]

# Possible values for each column
genders = ["Male", "Female", "Other"]
items = ["T-Shirt", "Laptop", "Running Shoes", "Handbag", "Smartphone", "Dress", "Wristwatch", "Backpack", "Gaming Console", "Perfume","kurta","sweater","jeans","sunglasses"]
categories = ["Clothing", "Electronics", "Footwear", "Accessories", "Beauty", "Bags"]
locations = ["New York", "Los Angeles", "Chicago", "Houston", "San Francisco", "Miami", "Seattle", "Boston", "Denver", "Las Vegas"]
sizes = ["S", "M", "L", "XL", "N/A"]
colors = ["Red", "Blue", "Black", "White", "Gray", "Pink", "Gold", "Silver","Green","Purple","Yellow","Orange","Olive","Grey"]
seasons = ["Winter", "Spring", "Summer", "Fall", "All Seasons"]
subscription_statuses = ["Active", "Inactive"]
payment_methods = ["Credit Card", "Debit Card", "PayPal", "Apple Pay", "Google Pay", "Digital Wallet"]
shipping_types = ["Standard", "Express","Free ship","Nextday","Store pickup"]
promo_codes = ["SUMMER20", "SPRING15", "TECH10", "BEAUTY5", "None"]
preferred_payment_methods = ["Credit Card", "Debit Card", "PayPal", "Apple Pay", "Google Pay"]
frequency_of_purchases = ["Weekly", "Bi-Weekly", "Monthly", "Quarterly", "Yearly"]

# Generate sample data for 3000 customers
sample_data_3000 = []
for i in range(1, 3001):
    age = random.randint(18, 60)
    gender = random.choice(genders)
    item = random.choice(items)
    category = random.choice(categories)
    purchase_amount = round(random.uniform(10, 1000), 2)
    location = random.choice(locations)
    size = random.choice(sizes)
    color = random.choice(colors)
    season = random.choice(seasons)
    review_rating = round(random.uniform(1, 5), 1)
    subscription_status = random.choice(subscription_statuses)
    payment_method = random.choice(payment_methods)
    shipping_type = random.choice(shipping_types)
    discount_applied = random.choice(["Yes", "No"])
    promo_code_used = random.choice(promo_codes) if discount_applied == "Yes" else "None"
    previous_purchase = random.choice(["Yes", "No"])
    preferred_payment_method = random.choice(preferred_payment_methods)
    frequency_of_purchase = random.choice(frequency_of_purchases)

    sample_data_3000.append([
        i, age, gender, item, category, purchase_amount, location, size, color, season,
        review_rating, subscription_status, payment_method, shipping_type, discount_applied,
        promo_code_used, previous_purchase, preferred_payment_method, frequency_of_purchase
    ])

# Convert to DataFrame
df_sample_3000 = pd.DataFrame(sample_data_3000, columns=columns)

# Save to CSV
df_sample_3000.to_csv("sample_data.csv", index=False)

print("CSV file with 3000 records has been created successfully!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("sample_data.csv")

# Convert boolean columns to categorical
df["Discount Applied"] = df["Discount Applied"].astype(str)
df["Previous Purchase"] = df["Previous Purchase"].astype(str)

# Emerging trends: Most purchased items
plt.figure(figsize=(12, 6))
sns.countplot(y=df["Item Purchased"], order=df["Item Purchased"].value_counts().index, palette="coolwarm")
plt.title("Top Purchased Items")
plt.xlabel("Count")
plt.ylabel("Item Purchased")
plt.show()

# Customer preferences: Preferred payment methods
plt.figure(figsize=(8, 5))
sns.countplot(x=df["Preferred Payment Method"], order=df["Preferred Payment Method"].value_counts().index, palette="viridis")
plt.title("Preferred Payment Methods")
plt.xlabel("Payment Method")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

# Seasonal buying preferences
plt.figure(figsize=(10, 5))
sns.countplot(x=df["Season"], order=df["Season"].value_counts().index, palette="pastel")
plt.title("Purchases by Season")
plt.xlabel("Season")
plt.ylabel("Count")
plt.show()

# Average purchase amount by category
plt.figure(figsize=(10, 5))
sns.barplot(x=df["Category"], y=df["Purchase Amount"], estimator=sum, palette="muted")
plt.title("Total Purchase Amount by Category")
plt.xlabel("Category")
plt.ylabel("Total Purchase Amount")
plt.xticks(rotation=45)
plt.show()
```

```python
# @title Gender that has highest number of purchases

gender_counts = df["Gender"].value_counts()
plt.figure(figsize=(3, 3))
sns.barplot(x=gender_counts.index, y=gender_counts.values, palette="coolwarm")
plt.title("Number of Purchases by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Purchases")
plt.show()
```

```python
# @title Overall distribution of customer ages

plt.figure(figsize=(8, 5))
sns.histplot(df["Age"], bins=20, kde=True, color="blue")
plt.title("Age Distribution of Customers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# @title How does purchse amount vary across diffrent product categories

plt.figure(figsize=(10, 6))
df.groupby("Category")["Purchase Amount"].mean().sort_values().plot(kind='bar', color='orange')
plt.xlabel("Categories")
plt.ylabel("Average Purchase Amount")
plt.title("Average Purchase Amount Across Different Categories")
plt.xticks(rotation=45)
plt.show()


```

```python
# @title Finding most commonly purchased items in each category

import pandas as pd
category_item_counts = df.groupby(['Category', 'Item Purchased'])['Item Purchased'].count().reset_index(name='Count')
most_frequent_items = category_item_counts.loc[category_item_counts.groupby('Category')['Count'].idxmax()]
print(most_frequent_items)
```

```python
# @title Specific seasons or months where customer spending is significantly high

plt.figure(figsize=(10, 5))
sns.barplot(x=df["Season"], y=df["Purchase Amount"], estimator=sum, palette="rocket")
plt.title("Total Customer Spending by Season")
plt.xlabel("Season")
plt.ylabel("Total Spending")
plt.xticks(rotation=45)
plt.show()
```

```python
# @title Average rating given by customer for each product category

avg_ratings = df.groupby('Category')['Review Rating'].mean().reset_index()

# Display the average ratings for each category
print("Average Ratings for Each Product Category:")
print(avg_ratings)

avg_ratings = df.groupby('Category')['Review Rating'].mean().reset_index()

# Plotting the bar plot
plt.figure(figsize=(8, 6))
plt.bar(avg_ratings['Category'], avg_ratings['Review Rating'], color='skyblue')
plt.title('Average Rating for Each Product Category')
plt.xlabel('Product Category')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()


```

```python
# @title Most popular payment method used

most_popular = df["Payment Method"].value_counts().idxmax()

print(f"The most popular payment method is: {most_popular}")
payment_counts = df['Payment Method'].value_counts()

```

```python
# Create the line plot
plt.figure(figsize=(4, 4))
sns.lineplot(x=payment_counts.index, y=payment_counts.values, marker='o', color='skyblue')
plt.title('Popularity of Payment Methods')
plt.xlabel('Payment Method')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.grid(True, linestyle='--', alpha=0.5)  # Add a grid for better visualization
plt.show()
```

```python
# @title How does the frequency of purchase vary across different age groups


bins = [18, 25, 35, 45, 55, 65]  # Adjust as needed
labels = ["18-24", "25-34", "35-44", "45-54", "55-64"]
df["Age Group"] = pd.cut(df["Age"], bins=bins, labels=labels, right=False)

age_group_counts = df["Age Group"].value_counts().sort_index()

print("\nPurchase frequency per age group:")
print(age_group_counts)

plt.figure(figsize=(8, 5))
age_group_counts.plot(kind="bar", color="skyblue", edgecolor="black")
plt.xlabel("Age Group")
plt.ylabel("Number of Purchases")
plt.title("Purchase Frequency Across Age Groups")
plt.xticks(rotation=45)
plt.show()
```

```python
# @title  Analysis on shipping type preffered

shipping_preference = df.groupby(['Category', 'Shipping Type']).size().reset_index(name='Count')

# Find the preferred shipping type for each product category
preferred_shipping = shipping_preference.loc[shipping_preference.groupby('Category')['Count'].idxmax()]

# Display the result
print(preferred_shipping)
```
